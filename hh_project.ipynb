```python
#️импортируем нужные библиотеки
import sqlite3
import pandas as pd
#from pandas_profiling import ProfileReport

```

подключение к существующей бд sqlite, либо создание нового файла бд и автоматическое подключение к нему

```python
#️подключение к бд
con = sqlite3.connect('hh_project.db', timeout=10)
cur = con.cursor()
```

```python
#️загрузка необходимых столбцов в dataframe
df = pd.read_excel('vacancies_data_for_analysis.xlsx', usecols=['id', 'name', 'area.id', 'address.metro.station_name', 'salary.from', 'salary.to',
'address.raw', 'experience.name', 'schedule.name', 'employment.name', 'employer.id', 'alternate_url', 'created_at'])
```

```python
#️загружаете вашу таблицу в бд, указав название таблицы. 
#таблица в базе данных создается в таком случае автоматически если ее нет
#если таблица с таким именем уже есть, то можно указать один из агрументов: 
#if_exists = 'replace' - пересоздаст таблицу (удалит существующую и создаст новую со вставкой ваших данных)
#if_exists = 'append' - добавит в текущую таблицу ваши данные (только если они подходят по форматам, это рабоает как insert into table)
df.to_sql(con=con, name='vacancies_data_for_analysis', index=False, if_exists = 'replace')
```

```python
#️считывание данных из таблицы (проверка)
data_test = cur.execute('select * from vacancies_data_for_analysis')
con.commit()
cur.fetchall()
```

```python
#загрузка справочников
df_area = pd.read_excel('area.xlsx')
df_employer = pd.read_excel('employer.xlsx')
```
